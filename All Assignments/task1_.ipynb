Q.1) What is the difference between "Love" and "love" in NLP?

Case 2: Question Answering

Q.4) Explain the difference between stemming and lemmatization

--> Stemming and lemmatization are both used to reduce words to their base form, but they work differently.

Stemming is a simple and fast method that cuts words without checking if the result is a real word. For example, “running” may become “runn”.

On the other hand, lemmatization is more accurate because it converts the word into a proper dictionary form. For example, “running” becomes “run”. Stemming is faster but less correct, while lemmatization is slower but gives better results.

Implement the following function:
def preprocess_text(text):

Required Features
Your function must include:
1. Remove numbers Example: "I have 2 dogs" → "I have dogs"
2. Remove extra spaces Example: "This is good" → "This is good"
3. Handle repeated characters Example: "soooo goooood!!!" → "so good"
4. Convert text to lowercase Example: "WOW!!! This is GREAT!!!" → "wow this is great"
5. Remove very short tokens
○ Remove words with length ≤ 2
○ Exception: keep meaningful words like "no", "not"
6. Remove URLs and email-like patterns Example: "Visit http://example.com now" → "visit now"

```python
import re

def preprocess_text_pipeline(text):

    print("Original Text :", text)
    print("=" * 120)

    # Step 1: Convert to lowercase
    before = text
    text = text.lower()
    print("Step 1: Lowercase")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 2: Remove URLs
    before = text
    text = re.sub(r'http\S+|www\S+', '', text)
    print("Step 2: Remove URLs")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 3: Remove Emails
    before = text
    text = re.sub(r'\S+@\S+', '', text)
    print("Step 3: Remove Emails")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 4: Remove Numbers
    before = text
    text = re.sub(r'\d+', '', text)
    print("Step 4: Remove Numbers")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 5: Handle Repeated Characters
    before = text
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    print("Step 5: Handle Repeated Characters")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 6: Remove Special Characters
    before = text
    text = re.sub(r'[^a-z\s]', '', text)
    print("Step 6: Remove Special Characters")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 7: Remove Extra Spaces
    before = text
    text = re.sub(r'\s+', ' ', text).strip()
    print("Step 7: Remove Extra Spaces")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 8: Remove Short Words (≤2 except 'no', 'not')
    before = text
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]
    text = " ".join(words)
    print("Step 8: Remove Short Words")
    print("Before :", before)
    print("After  :", text)
    print("-" * 120)

    # Step 9: Tokenization
    tokens = text.split()
    print("Step 9: Tokens")
    print("Tokens :", tokens)
    print("=" * 120)

    return tokens, text

text = "WOW!!! This is soooo goooood!!! Visit http://example.com now. I have 2 dogs!! Email me at test@gmail.com"

preprocess_text_pipeline(text)
```

**Task 4: Token Analytics**


For each processed sentence, compute: Total number of tokens Number of unique tokens Average token length

Analysis Questions Which sentence had the most noise? Which sentence retained the most meaningful tokens after cleaning?

```python
# Sentences
# -----------------------------
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("\n--- Token Analytics ---\n")

results = []

# -----------------------------
# Compute Metrics
# -----------------------------
for i, text in enumerate(sentences, 1):

    tokens, cleaned = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    avg_length = round(sum(len(t) for t in tokens) / total_tokens, 2) if total_tokens > 0 else 0

    results.append({
        "sentence_no": i,
        "original": text,
        "tokens": tokens,
        "total": total_tokens,
        "unique": unique_tokens,
        "avg_len": avg_length
    })

    print(f"Sentence {i}:")
    print("Original Text:        ", text)
    print("Cleaned Tokens:       ", tokens)
    print("Total Tokens:         ", total_tokens)
    print("Unique Tokens:        ", unique_tokens)
    print("Average Token Length: ", avg_length)
    print("-" * 50)


# -----------------------------
# Analysis Answers
# -----------------------------

# Most noise → sentence with biggest reduction (original length vs tokens)
most_noise = min(results, key=lambda x: x["total"])

# Most meaningful → sentence with highest unique tokens
most_meaningful = max(results, key=lambda x: x["unique"])

print("\n--- Analysis ---\n")

print("Sentence with MOST noise:")
print(f"Sentence {most_noise['sentence_no']}: {most_noise['original']}")

print("\nSentence with MOST meaningful tokens:")
print(f"Sentence {most_meaningful['sentence_no']}: {most_meaningful['original']}")
```

**Task 5: Frequency Analysis**


Combine all tokens and use: from collections import Counter

Identify: Top 10 most frequent words and Top 5 least frequent words

```python
from collections import Counter
import re

# -----------------------------
# Preprocessing Function
# -----------------------------
def preprocess_text_pipeline(text):

    # Convert text to lowercase
    text = text.lower()

    # Remove URLs (http / www links)
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers (digits)
    text = re.sub(r'\d+', '', text)

    # Remove repeated characters (e.g., loooove → love)
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove special characters and emojis (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove short words (length ≤ 2), but keep 'no' and 'not'
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    # Final tokens (list of words)
    tokens = words

    # Return tokens and cleaned sentence
    return tokens, text


# -----------------------------
# Input Sentences
# -----------------------------
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


# -----------------------------
# Combine all tokens from sentences
# -----------------------------
all_tokens = []  # empty list to store all tokens

for text in sentences:

    # Get tokens from preprocessing function
    tokens, _ = preprocess_text_pipeline(text)

    # Add tokens to the main list
    all_tokens.extend(tokens)


# -----------------------------
# Count word frequencies using Counter
# -----------------------------
counter = Counter(all_tokens)


# -----------------------------
# Get Top 10 Most Frequent Words
# -----------------------------
top_10 = counter.most_common(10)


# -----------------------------
# Get Top 5 Least Frequent Words
# -----------------------------
# Sort words based on frequency (ascending order)
least_5 = sorted(counter.items(), key=lambda x: x[1])[:5]


# -----------------------------
# Print Results
# -----------------------------
print("\n--- Token Frequency Analysis ---\n")

# Print most frequent words
print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word}: {freq}")

# Print least frequent words
print("\nTop 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word}: {freq}")
```

**Task 6: Build Full Pipeline**


def full_pipeline(text_list): pass

Expected Output { "tokens": [...], "clean_sentences": [...] }

```python
# -----------------------------
# Full Pipeline Function
# -----------------------------
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text_pipeline(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# -----------------------------
# Run Pipeline
# -----------------------------
result = full_pipeline(sentences)


# -----------------------------
# Clean Output Display
# -----------------------------
print("\n--- FINAL PIPELINE OUTPUT ---\n")

# Print tokens nicely
print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for i, sent in enumerate(result["clean_sentences"], 1):
    print(f"{i}. {sent}")
```

```python
def preprocess_text(text):

    #  Handle None or invalid input FIRST
    if not text or not isinstance(text, str):
        return [], ""

    # Now safe to process
    text = text.lower()

    import re
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    cleaned = " ".join(words)
    tokens = cleaned.split() if cleaned else []

    return tokens, cleaned

# -----------------------------
# Test Function for Error Handling
# -----------------------------
def test_error_handling():

    test_cases = {
        "Empty String": "",
        "Only Emojis": "😍😍😍😍",
        "Only Numbers": "1234567890",
        "Spaces Only": "     ",
        "None Value": None
    }

    print("\n--- Task 7: Error Handling Results ---\n")

    all_passed = True  # flag to track overall success

    for name, input_text in test_cases.items():

        # Run preprocessing
        tokens, cleaned = preprocess_text(input_text)

        # Check if handled correctly (expect empty output)
        if tokens == [] and cleaned == "":
            status = "Handled"
        else:
            status = "Not Handled"
            all_passed = False

        # Print formatted output
        print(f"Test Case: {name}")
        print(f"  Input:  '{input_text}'")
        print(f"  Output: '{cleaned}'")
        print(f"  Status: {status}")
        print("-" * 40)

    # Final result
    if all_passed:
        print("\n All Edge Cases Passed! The Preprocessing Engine is robust.")
    else:
        print("\n Some test cases failed. Please check the pipeline.")

test_error_handling()
```
